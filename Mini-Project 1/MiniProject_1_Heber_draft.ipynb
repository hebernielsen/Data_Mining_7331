```python
## TRAINING AND TESTING
```

```python
from sklearn.model_selection import ShuffleSplit

# predict the X and y data:
if 'Ponderal_index' in df_newlite:
    y = df_newlite['Exposed'].values # get the labels we want
    del df_newlite['Exposed'] # get rid of the class label
    X = df_newlite.values # use everything else to predict!

    ## X and y are now numpy matrices, by calling 'values' on the pandas data frames we
    #    have converted them into simple matrices to use with scikit learn
    
    
# to use the cross validation object in scikit learn, we need to grab an instance
#    of the object and set it up. This object will be able to split our data into 
#    training and testing splits
num_cv_iterations = 3
num_instances = len(y)
cv_object = ShuffleSplit(n_splits=num_cv_iterations,
                         test_size  = 0.2)
                         
print(cv_object)
```

```python
## LOGISTIC REGRESSION
```

```python
# run logistic regression and vary some parameters
from sklearn.linear_model import LogisticRegression
from sklearn import metrics as mt

# Create a reusable logisitic regression object
# Set up the object with different learning parameters and constants
lr_clf = LogisticRegression(penalty='l2', C=1.0, class_weight=None) # This is the object

# Use the cv_object that we set up before to iterate through the 
#    different training and testing sets. Each time reuse the logisitic regression 
#    object, but train it on different data each time it is used.
# Every time the following code is run
#   it randomly creates a new training and testing set, 
#   so accuracy will be different each time

iter_num=0
# the indices are the rows used for training and testing in each iteration
for train_indices, test_indices in cv_object.split(X,y): 
    # Create new variables 
    X_train = X[train_indices]
    y_train = y[train_indices]
    
    X_test = X[test_indices]
    y_test = y[test_indices]
    
    # train the reusable logisitc regression model on the training data
    lr_clf.fit(X_train,y_train)  # train object
    y_hat = lr_clf.predict(X_test) # get test set predictions

    # Get the accuracy and confusion matrix for this iterations of training/testing
    acc = mt.accuracy_score(y_test,y_hat)
    conf = mt.confusion_matrix(y_test,y_hat)
    print("====Iteration",iter_num," ====")
    print("accuracy", acc )
    print("confusion matrix\n",conf)
    iter_num+=1
    
## AS SEEN IN THE OUTPUT, THIS IS TOO ACCURATE, WITH VALUES OF 1.0
```

```python
## Redo the Logistic Regression using the other two code chunks we were given (more compact code).
## Expect similar output as above (matrices will of course be slightly different)

## Chunk #2 - intermediate compactness
for iter_num, (train_indices, test_indices) in enumerate(cv_object.split(X,y)):
    lr_clf.fit(X[train_indices],y[train_indices])  # train object
    y_hat = lr_clf.predict(X[test_indices]) # get test set precitions

    # print the accuracy and confusion matrix 
    print("====Iteration",iter_num," ====")
    print("accuracy", mt.accuracy_score(y[test_indices],y_hat)) 
    print("confusion matrix\n",mt.confusion_matrix(y[test_indices],y_hat))

## AS SEEN IN THE OUTPUT, THIS IS TOO ACCURATE, WITH VALUES OF 1.0
```

```python
## Chunk #3 - very compact
## Apparently this also can help with parallelism
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(lr_clf, X, y=y, cv=cv_object) 
print(accuracies)

## AS SEEN IN THE OUTPUT, THIS IS TOO ACCURATE, WITH VALUES OF 1.0
```
